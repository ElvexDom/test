```python
import os

PATH = r'.env'

Nfiles = 0
Ndirs = 0

for root, dirnames, filenames in os.walk(PATH):
    Ndirs += len(dirnames)
    Nfiles += len(filenames)

print("Ndirs", Ndirs)
print("Nfiles",Nfiles)
```
